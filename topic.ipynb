```python
#outlier detection
num_col=['Age','TSH_Level','T3_Level','T4_Level','Nodule_Size']
sns.boxplot(data=df[num_col])
plt.show()
```

```python
Q1=np.percentile(df['Age'],25,method='midpoint')
Q2=np.percentile(df['Age'],50,method='midpoint')
Q3=np.percentile(df['Age'],75,method='midpoint')
print(Q1)
print(Q2)
print(Q3)

IQR= Q3-Q1
print(IQR)

upper_limit=Q3 + 1.5 * IQR
lower_limit=Q1 - 1.5 * IQR
print(upper_limit)
print(lower_limit)
```

```python
outlier=[]
for x in df['Age']:
    if (x>upper_limit) or (x<lower_limit):
        outlier.append(x)
outlier
     
```
