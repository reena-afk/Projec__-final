```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('thyroid_cancer_risk_data.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diagnosis',y='TSH_Level',data=df)
plt.title('TSH Level Distribution by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('TSH Level') 
plt.show()
```

```python
sns.boxplot(x='Diagnosis',y='Nodule_Size',data=df)
plt.title('Nodule Size Distribution by Diagnosis')
plt.show()
```

```python
sns.countplot(x='Family_History',hue='Diagnosis',data=df)
plt.title('Family History of Thyroid Cancer by Diagnosis')
plt.show()
```

```python
sns.countplot(x='Iodine_Deficiency',hue='Diagnosis',data=df)
plt.title('Iodine Deficiency by Diagnosis')
plt.show()
```

```python
sns.countplot(x='Radiation_Exposure',hue='Diagnosis',data=df)
plt.title('Radiation Exposure by Diagnosis')
plt.show()
```

```python
sns.countplot(x='Family_History',hue='Diagnosis',data=df)
plt.title('Family History of Thyroid Cancer by Diagnosis')
plt.show()
```

```python
sns.countplot(x='Smoking',hue='Diagnosis',data=df)
plt.title('Smoking Status by Diagnosis')
plt.show()
```

```python
sns.set_style('whitegrid')
target='Thyroid_Cancer_Risk'

cat_col=df.select_dtypes(include='object').columns
for col in cat_col:
    if col!=target:
        plt.figure(figsize=(8, 4))
        sns.countplot(x=col,hue=target,data=df)
        
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.legend(title=target)
        plt.show()
```

```python
sns.set_style('whitegrid')
cols=['Age','TSH_Level','T3_Level','T4_Level','Nodule_Size']
for col in cols:
    plt.figure(figsize=(8, 4))
    sns.barplot(x='Thyroid_Cancer_Risk',y=col,data=df)
    plt.title(f'{col} vs Thyroid Cancer Risk')
    plt.xlabel('Thyroid Cancer Risk')
    plt.ylabel(col)
    plt.show()
```
